```python
import torch
```

```python
x = torch.rand(10, 784)
w = torch.rand(784, 10)
b = torch.rand(10)

y_true = torch.rand(10, 10)
print(y_true)


def sigmoid(x):
    return 1 / (1 + torch.exp(-x))

def gradients(x, y, y_hat, z):
    print(y_hat)
    print(y)
    dL_dZ = y_hat - y
    dZ_dW = x
    dL_dW = dZ_dW.T @ dL_dZ
    dL_db = (dL_dZ).sum(dim=0)
    return dL_dW, dL_db
```

```python

def train_perceptron(x, w, b, y_true, mu):
    #forward propagation
    z = x @ w + b
    y_hat = sigmoid(z)
    #print("testtest")
    print(y_hat)

    dL_dW, dL_db = gradients(x, y_true, y_hat, z)
    w = w - mu * dL_dW
    b = b - mu * dL_db
    return w, b

```

```python
mu = 0.1
w, b = train_perceptron(x, w, b, y_true, mu)

#print(w)
#print(b)
print(w.shape)
print(b.shape)
```
